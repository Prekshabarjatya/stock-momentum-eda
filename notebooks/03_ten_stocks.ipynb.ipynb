```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 10 NIFTY 50 stocks - different sectors
tickers = [
    # Banking
    "HDFCBANK.NS",
    "ICICIBANK.NS",
    "KOTAKBANK.NS",
    
    # IT
    "TCS.NS",
    "INFY.NS",
    "WIPRO.NS",
    
    # Energy
    "RELIANCE.NS",
    "ONGC.NS",
    
    # FMCG
    "HINDUNILVR.NS",
    
    # Auto
    "MARUTI.NS"
]

print(f"Total stocks: {len(tickers)}")
print("Sectors covered: Banking, IT, Energy, FMCG, Auto")
```

```python
raw_data = yf.download(tickers, start = "2020-01-01", end = "2025-01-01")

close_prices  = raw_data["Close"]

print("Shape:", close_prices.shape)
print("\nColumns:", close_prices.columns.tolist())

close_prices.head()
```

```python
df_long =  close_prices.reset_index().melt(
    id_vars = "Date",
    var_name = "Tickers",
    value_name = "Close"
)

print("Shape: ", df_long.shape)
print("\nFirst 10 rows: ")
df_long.head(-10)
```

```python
df_long.to_csv("../data/ten_stocks_long.csv", index = False)
close_prices.to_csv("../data/ten_stocks_wide.csv")

print("Saved!")
```

```python
#Stock wise avg close price

avg_stock_price = df_long.groupby("Tickers")["Close"].mean().round(2)

print("Average Close Price (2020-2025): ")
for ticker, price in avg_stock_price.sort_values(ascending = False).items():
    print(f"{ticker} : Rs. {price}")
```
